```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(6, 4))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--',color='pink')
plt.style.use('dark_background')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.grid()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset("iris")

# Define a custom color palette
custom_palette = {"setosa": "blue", "versicolor": "green", "virginica": "purple"}

# Create a Violin Plot with custom colors
sns.violinplot(x=iris["species"], y=iris["sepal_length"], palette=custom_palette)

plt.title('Custom Colored Violin Plot of Sepal Length')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

```python
import numpy as np
from statsmodels.distributions.empirical_distribution import ECDF

ecdf = ECDF(iris["sepal_length"])
plt.plot(ecdf.x, ecdf.y)
plt.title('ECDF Plot of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Cumulative Probability')
plt.show()
```
